```python
# Importing necessary libraries
import pandas as pd
import numpy as np 
import ast
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Use pandas read_csv function to load csv as DataFrame
df = pd.read_csv('../input/dataset/twitter_HOTD_DA_WORKSHEET4A.csv')
df
```

# PROBLEM 1
Tokenize the string representations provided in the **tokens** column of the DataFrame using TF-IDF from sklearn. Then print out the TF-IDF of the first row of the DataFrame.

```python
# Convert string representation of a list into a list of strings
text = []
for r in df['tokens']:
    res = ast.literal_eval(r)
    if(' '.join(res).lower() == ''):
        print(r)
    text.append(' '.join(res).lower())
# Print the end result
text
```

```python
tf_idf = TfidfVectorizer()
text_tf = tf_idf.fit_transform(text)
text_tf = tf_idf.transform(text)
print("n_samples: %d, n_features: %d" % text_tf.shape)
text_tf=text_tf.toarray()
print(text_tf)
```

# PROBLEM 2
Find the top-5 most similar tweets to the tweet with index `7558` using cosine similarity between the TF-IDF vectors.

```python
result=cosine_similarity(text_tf, dense_output=True)
result=np.argsort(result)
```

```python
for i in range(5):
    index=result[7558][i]
    print(text[index])
    print(df.iloc[index])
```
