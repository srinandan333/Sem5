```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("../input/mi-assignment/MI_Assignment1_2_Online Retail.xlsx")
```

```python
df.info()
```

```python
from statistics import mean, mode

m_d=mode(df['Description'])
m_d
```

```python
from pandas.core.algorithms import mode
df['Description'].fillna(m_d,inplace=True)
a=[]
for i in df["CustomerID"]:
  if not np.isnan(i):
    a.append(i)
df["CustomerID"].fillna(float(mode(a)),inplace=True)
for i in df.index:
  if df['Quantity'][i]<0:
    df['Quantity'][i]=df['Quantity'][i]*-1
```

```python
ind=0
for i in range(len(df)):
    if df['InvoiceDate'][i]> pd.Timestamp(2011,9,1):
        ind=i
        break
```

```python
df[ind+1:]
```

```python
df_t=df[:][:ind]
df_t
```

```python
after=[]
for i in df['CustomerID'][ind+1:]:
    after.append(i)
len(after)    
after=np.array(after)
after=np.unique(after)
```

```python
lis=[0]*len(df_t)
for i in range(len(df_t)):
    if df_t['CustomerID'][i] in after:
        lis[i]="Not Churn"
    else:
        lis[i]="Churn"
```

```python
df_t['Target']=lis
```

```python
df_t['Target']
```

```python
df_t
```

```python
del df_t['InvoiceDate']
del df_t['Description']
del df_t['Country']
del df_t['StockCode']
del df_t['InvoiceNo']
df_t=df_t.dropna()
```

```python
x = df_t.iloc[:,0:-1].values
y = df_t.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
from sklearn.svm import SVC
model =SVC()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_pred,y_test)
```

```python
a1=accuracy_score(y_pred,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```
