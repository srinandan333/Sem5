```python
import pygad.gann
import numpy as np
import pygad.nn
import pygad
```

```python
input = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
output = np.array([0, 1, 1, 0])
```

```python
inst_gann = pygad.gann.GANN(hidden_activations=["sigmoid"], output_activation="softmax",  num_neurons_hidden_layers=[2], num_neurons_output=2, num_neurons_input=2, num_solutions=10)
```

```python
def func_gen(inst_ga):
    global inst_gann
    mat_pop = pygad.gann.population_as_matrices(population_networks=inst_gann.population_networks, population_vectors=inst_ga.population)
    inst_gann.update_population_trained_weights(population_trained_weights=mat_pop)
    print("Gen = {generation}".format(generation=inst_ga.generations_completed))
    print("Acc = {fitness}".format(fitness=inst_ga.best_solution()[1]))
```

```python
population_vectors = pygad.gann.population_as_vectors(population_networks=inst_gann.population_networks)
```

```python
def func_fit(result, res_idx):
    global inst_gann, input, output
    pred = pygad.nn.predict(last_layer=inst_gann.population_networks[res_idx], data_inputs=input)
    right_pred = np.where(pred == output)[0].size
    fitness = (right_pred/output.size)*100
    return fitness
```

```python
inst_ga = pygad.GA(initial_population=population_vectors.copy(), num_generations=50, on_generation=func_gen, mutation_num_genes=2, num_parents_mating=2, fitness_func=func_fit)
```

```python
inst_ga.run()
```

```python
result, fitness, res_idx = inst_ga.best_solution()
print("Solution: ",result)
print("Accuracy: ", fitness)
```
